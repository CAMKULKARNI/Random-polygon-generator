```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for _ in range(10):
    print(random.vonmisesvariate(mu=np.pi, kappa=0.6))
```

```python
n = int(input("Enter the number of vertices : "))
r = np.random.rand() * 1000
theta = np.array(sorted(random.vonmisesvariate(mu=np.pi / 2, kappa=1.319) for i in range(n)))
theta = np.append(theta, theta[0])
x = r * np.cos(theta)
y = r * np.sin(theta)
coords = list(zip(x, y))

plt.plot(x, y)
plt.show()
```
