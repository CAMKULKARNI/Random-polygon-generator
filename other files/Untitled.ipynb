```python
import numpy as np
import random 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
radii = sorted([np.random.rand() * 100 for _ in range(5)])
print(radii)
```

```python
n = int(input("Enter the number of vertices : "))
theta = np.array(sorted(random.vonmisesvariate(mu=np.pi, kappa=1) for i in range(n)))
```

```python
x = []
y = []
for i in range(n):
    r = random.choice(radii)
    x.append(r * np.cos(theta[i]))
    y.append(r * np.sin(theta[i]))
    
x.append(x[0])
y.append(y[0])

plt.figure()
plt.plot(x, y)
plt.show()
```
